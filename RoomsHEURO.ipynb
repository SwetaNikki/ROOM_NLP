```python
def charToString(documents):
    document = []
    #char=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    a = ""
    for doc in documents:
        if doc==" ":
            document.append(a)
            a = ""
        elif doc==".":
            document.append(a)
            a = "."
        elif doc==",":
            document.append(a)
            a = ","
        else:
            a = a + doc
    return document
 
```

```python
def separate_With_connected_link(document):
    attach = ["connected", "next", "adjacent"]
    class my_dictionary(dict):  
        # __init__ function  
        def __init__(self):  
            self = dict()  
          
        # Function to add key:value  
        def add(self, key, value):  
            self.setdefault(key, [])
            self[key].append(value)  
  
    room_type = my_dictionary()

    for i in range(len(document)): 
        for j in range(len(attach)):
            if ((document[i]=="next") or (document[i]=="adjacent")):
                room_type.add(document[i-2], document[i+2])
            elif document[i]=="connected":
                # this1 , this2 and this3 are connected
                #  i-6 i-5  i-4  i-3 i-2   i-1    i
                # room   link
                # this1 this2
                # this2 this3
                # this3 this1
                if (document[i-5]==","):
                    room_type.add(document[i-6], document[i-4])
                    room_type.add(document[i-4], document[i-2])
                    room_type.add(document[i-2], document[i-6])
                elif (document[i-5]=="."):
                    room_type.add(document[i-4], document[i-2])
                    room_type.add(document[i-2], document[i-4])
            
    return room_type
```

```python
for i in range(len(data)):
    print("DOCUMENT = ", i)
    print()
    documents = data[i]['document']
    document = charToString(documents)
    room_type = separate_With_connected_link(document)
    dict_to_table(room_type)
```
